# Mini-Project 1 - Interest calculator

Write a program that calculates compound interest.

1. Your program should ask the user how much they wish to deposit in the first year, at what interest rate and then should calculate the savings after the first year. The interest should compound annually.
2. It should then ask how much the user wishes to deposit in the second year, add this on to the first year saving and calculate interest after the second year. 
3. Structure the program in a while loop so that it repeats step 2 for further years until the user enters “0” to stop.
