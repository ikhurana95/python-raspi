# Mini-Project 2 - Storyteller

Write a program that lets you build sentences from a list of words (come up with your own list of words).

	words = ["once", "the", "a", "and", "time", "then", "rabbit", "jumped", "fence", "upon"]

You should be able to exit the program by entering _quit_.

Example output:

	Words at your disposal:
	1. once
	2. the
	3. a
	4. and
	5. time
	6. then
	7. rabbit
	8. jumped
	9. fence
	10. upon

	>> 1

	once

	>> 10

	Your sentence:
	once upon

	Choose your word:
	>> 3

	Your sentence:
	once upon a

	Choose your word:
	>> 5

	>> 3

	once upon a time a

	Choose your word:
	quit

_Note 1:_ Sometimes you want to loop through a list whilst also keeping track of the _index_ of the list element. You can do this by combining the _range_ and _len_ functions:

	my_list = [7, 8, 1, 3, 4]

	for i in range(len(my_list)):
		print("List element " + str(i) + ": " + str(my_list[i]))

_Note 2:_ Remember that if you want to request a number from a user, you have to use _int_ and _input_, like this:

	str = input("Enter a number: ") # This is a string
	num = int(str) # This is an integer
